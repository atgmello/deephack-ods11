## Group expenses by themes or subthemes

```python
themed_df = expenses.group_by_theme(df)
themed_df.head(20)
```

## Building historical data of expenses

### Historical data of expenses by theme

```python
years = range(2010, 2017)
city = "araraquara"
themes = ["EDUCAÇÃO", "SAÚDE"]

historic_df = expenses.get_themed_historical(city, years, themes)
historic_df.head()
```

```python
historic_df.plot.line(figsize=(10, 10))
```
