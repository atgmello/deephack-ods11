```python
%%time
data_path = '../data/cdhu'
for opt in radm_arr:
    print('\nIniciando obtenção de dados para {}'.format(opt))
    file_path = uni.unidecode(opt.replace(" ", "").lower())+".csv"
    cdhudf.get_df(search=opt, category='radm').to_csv(os.path.join(data_path,file_path))
```

```python
file_path = '../data/cdhu'

df = pd.DataFrame()
files = [f for f in os.listdir(file_path) if f not in ['.ipynb_checkpoints','proc.csv','proc_nosp.csv']]
for f in files:
    df = df.append(pd.read_csv(os.path.join(file_path,f)))
```

```python
df.head()
```

```python
df = cdhudf.preproc_df(df, mun_col=True)
```

```python
df.head()
```

```python
df.insert(8,'Total de UHS + Total de CCs',df['Total de UHS']+df['Total de CCs'])
```

```python
file_name = 'proc.csv'
df.to_csv(os.path.join(file_path,file_name),index=False)
```

```python
# Filtrando o município de São Paulo
df = df[df['Município'] != 'São Paulo']
```
